# Latar Belakang:

Dalam upaya meningkatkan jumlah pendaftar, Universitas XYZ telah menggunakan berbagai saluran periklanan untuk menarik calon mahasiswa. Saluran yang digunakan meliputi:

1. Brosur
2. Sosialisasi
3. Spanduk Penerimaan
4. Platform Digital

Meskipun telah melakukan berbagai upaya, Universitas XYZ ingin memahami efektivitas masing-masing saluran periklanan tersebut untuk mengoptimalkan anggaran iklan dan strategi periklanan ke depan. 

Buatlah model machine learning yang dapat menganalisis data dari berbagai saluran periklanan untuk mengidentifikasi saluran mana yang paling efektif dalam menarik pendaftar.

# Tujuan Studi Kasus: 

Mengukur Efektivitas Periklanan dari berbagai saluran iklana untuk mengetahui saluran mana yang memberikan anggaran pengeluaran tertinggi.




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("../data/pmb_data_universitas_xyz.xlsx")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
# Distribusi Variabel Target (Status Pendaftaran)
plt.figure(figsize=(8, 6))
sns.countplot(x='Status Pendaftaran', data=data)
plt.title('Distribusi Status Pendaftaran')
plt.show()
```

```python
# Distribusi Fitur Numerik
data.hist(bins=15, figsize=(15, 10))
plt.suptitle('Distribusi Fitur Numerik')
plt.show()
```

```python
# Matriks Korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Matriks Korelasi')
plt.show()
```

```python
# Distribusi Saluran Iklan
plt.figure(figsize=(10, 6))
sns.countplot(y='Saluran Iklan', data=data, order=data['Saluran Iklan'].value_counts().index)
plt.title('Distribusi Saluran Iklan')
plt.show()
```

```python
# Analisis Saluran Iklan berdasarkan Status Pendaftaran
plt.figure(figsize=(12, 6))
sns.countplot(y='Saluran Iklan', hue='Status Pendaftaran', data=data)
plt.title('Analisis Saluran Iklan berdasarkan Status Pendaftaran')
plt.show()
```
